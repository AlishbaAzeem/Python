**Classify Movies Based on Ratings [Excellent, Good and Average]**

```python
data.columns
```

```python
def Rating(rating):
  if rating>=7.0:
    return 'Excellent'
  elif rating<7.0 and rating>=6:
    return 'Good'
  else:
    return 'Average'
```

```python
data['rating_cat']=data['Rating'].apply(Rating)
data.head()
```

**Find Unique Values From Genre**

```python
data.columns
```

```python
data['Genre']
```

```python
listl=[]
for value in data['Genre']:
  listl.append(value.split(','))
```

```python
listl
```

```python
one_d=[]
for item in listl:
  for iteml in item:
    one_d.append(iteml)
```

```python
one_d
```

```python
uni_list=[]
for item in one_d:
  if item not in uni_list:
    uni_list.append(item)
```
